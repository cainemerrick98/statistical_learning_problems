**1. It was mentioned in this chapter that a cubic regression spline with one knot at $\xi$, can be obtained using a basis of the form $x, x^{2}, x^{3}, (x - \xi)^{3}_{+}$ where $(x - \xi)^{3}_{+} = (x - \xi)^{3}$ if $(x - \xi)^{3}_{+} > 0$ and $0$ otherwise.**

We will now show that the functional form

$$
f(x) = \beta_{0} + \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3} + \beta_{4}(x - \xi)^{3}_{+}
$$

Is indeed a cubic regression spline regardless of the values of $\beta_{0},\beta_{1},\beta_{2},\beta_{3},\beta_{4}$

This question is all about proving that adding the truncated power basis function, $\beta_{4}(x - \xi)^{3}_{+}$, to the cubic polynomial results in a function that is continuous in the first and second derivatives at the knot $\xi$.

*(a) Find a cubic polynomial*

$$
f_{1}(x) = a_{1} + b_{1}x + c_{1}x^{2} + d_{1}x^{3}
$$

*such that $f(x) = f_{1}(x)$ for all $x\le \xi$. Express $a_{1}, b_{1}, c_{1}, d_{1}$ in terms of $\beta_{0}, \beta_{1},\beta_{2},\beta_{3},\beta_{4}$*

We need to find a cubic polynomial $f_{1}(x)$ such that,

$$
f(x) = \beta_{0} + \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3} + \beta_{4}(x - \xi)^{3}_{+}
$$
$$
= \beta_{0} + \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3} \text{ since } (x - \xi)^{3}_{+} = 0 \forall x \le \xi 
$$
$$
= a_{1} + b_{1}x + c_{1}x^{2} + d_{1}x^{3}
$$
$$
= f_{1}(x)
$$

Clearly, when $x\le\xi$ we have $f(x) = f_{1}(x)$ when,

$$
a_{1} = \beta_{0}
$$
$$
b_{1} = \beta_{1}
$$
$$
c_{1} = \beta_{2}
$$
$$
d_{1} = \beta_{3}
$$

*(b) Find a cubic polynomial*

$$
f_{2}(x) = a_{2} + b_{2}x + c_{2}x^{2} + d_{2}x^{3}
$$

*such that $f(x) = f_{2}(x)$ for all $x\gt \xi$. Express $a_{2}, b_{2}, c_{2}, d_{2}$ in terms of $\beta_{0}, \beta_{1},\beta_{2},\beta_{3},\beta_{4}$*

When $x>\xi$ we have $(x - \xi)^{3}_{+} = (x - \xi)^{3}$, therefore we need to find a cubic polynomial $f_{2}(x)$ such that,

$$
f(x) = \beta_{0} + \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3} + \beta_{4}(x - \xi)^{3}
$$
$$
= a_{2} + b_{2}x + c_{2}x^{2} + d_{2}x^{3}
$$
$$
= f_{2}(x)
$$

$$
\beta_{4}(x - \xi)^{3} = \beta_{4}(x - \xi)^{2}(x^{2} - 2x\xi + \xi^{2}) = \beta_{4}(x^{3} - 3x^{2}\xi + 3x\xi^{2} - \xi^{3})
$$
$$
f(x) = \beta_{0} + \beta_{1}x + \beta_{2}x^{2} + \beta_{3}x^{3} + \beta_{4}(x^{3} - 3x^{2}\xi + 3x\xi^{2} - \xi^{3})
$$
$$
f(x) = (\beta_{0} - \beta_{4}\xi^{3}) + (\beta_{1} +3\xi)x + (\beta_{2} - 3\beta_{4}\xi)x^{2} + (\beta_{3} -\beta_{4})x^{3}
$$

Equating coefficients we have,

$$
a_{1} = (\beta_{0} - \beta_{4}\xi^{3})
$$
$$
b_{1} = (\beta_{1} +3\beta_{4}\xi)
$$
$$
c_{1} = (\beta_{2} - 3\beta_{4}\xi)
$$
$$
d_{1} = (\beta_{3} -\beta_{4})
$$

We have now established that $f(x)$ is a piecewise polynomial

**2. Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of n points using the following formula:**

$$
\hat{g} = argmin_{g}(\sum_{i=0}^{n}(y_{i} - g(x_{i}))^{2} + \lambda \int[g^{m}(x)]^{2}dx)
$$

Where $g^{m}$ represents the mth derivative of g and $g^{0} = g$. Provide example sketches of $\hat{g}$ in each of the following scenarios.

The function above is composed of a loss function, $\sum_{i=0}^{n}(y_{i} - g(x_{i}))^{2}$, this is the function we want to minimise to make $g$ fit the data closely. It also has a penalty term, $\lambda \int[g^{m}(x)]^{2}dx$, depending on the value of $\lambda$ this term penalises a function for being too wiggly. The larger the value of $\lambda$ the more biased our model is and the less variance it will have - bias variance trade off!

To skecth $\hat{g}$ we will use matplotlib. We will also need to generate some data.

We will use,

$$
Y = g(X) + \epsilon
$$

With 

$$
X \sim U[0, 1]
$$
$$
\epsilon \sim N[0, 1]
$$

And,

$$
g(X) = 10 + 3x - 5x^{2} + 7x^{3}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epsilon = np.random.normal(0, 2, size=100)
X = np.random.uniform(0, 2, size=100)
Y = 10 + 0*X - 10*X**2 + 7*X**3 + epsilon
```

```python
x_r = np.arange(200) / 100
g_x = 10 + 0*x_r - 10*x_r**2 + 7*x_r**3
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data and true generating function g(X)')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
```

*(a) $\lambda = \infty$ and $m = 0$*

When $\lambda = \infty$ the first term can be ignored in the minimisation. As $m=0$ if $g^{(0)}=g(x)=0$ then the function will be minimised at $\hat{g}=0$.

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data, true generating function g(X), and g hat')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
ax.axhline(0, c='orange') #g hat
```

*(b) $\lambda = \infty$ and $m = 1$*

When $\lambda = \infty$ the first term can be ignored in the minimisation. As $m=1$ we have $g^{(1)}=g'(x)=0$ therefore $\hat{g} = c$ because the first derivative of a constant is zero. In this case c will be $\frac{1}{n}\sum_{i=0}^{n}y_{i}$.

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data, true generating function g(X), and g hat')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
ax.axhline(np.mean(Y), c='orange') #g hat
```

*(c) $\lambda = \infty$ and $m = 2$*

When $\lambda = \infty$ the first term can be ignored in the minimisation. As $m=2$ we have $g^{(2)}=g''(x)=0$ which mean $\hat{g} = ax + b$ because the second derivative of a linear equation is zero.

In this case $ax + b$ will be linear least squares line.

```python
import statsmodels.api as sm
```

```python
ols = sm.OLS(Y, sm.add_constant(X))
results = ols.fit()
params = results.params
```

```python
ols_line = params[0] + x_r*params[1]
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data, true generating function g(X), and g hat')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
ax.plot(x_r, ols_line, c='orange') #g hat
```

*(d) $\lambda = \infty$ and $m = 3$*

When $\lambda = \infty$ the first term can be ignored in the minimisation. As $m=2$ we have $g^{(3)}=g'''(x)=0$ which means $\hat{g} = ax^{2} + bx + c$ because the third derivative of a quadratic equation is zero.

In this case $ax^{2} + bx + c$ will be quadratic least squares line.

```python
X_X2 = np.hstack((X.reshape(100, 1), (X**2).reshape(100,1)))
ols = sm.OLS(Y, sm.add_constant(X_X2))
results = ols.fit()
params = results.params
params
```

```python
ols_quad = params[0] + x_r*params[1] + params[2]*x_r**2
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data, true generating function g(X), and g hat')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
ax.plot(x_r, ols_quad, c='orange') #g hat
```

*(e) $\lambda = 0$ and $m = 3$*

When $\lambda = 0$ the second term can be ignored. g(x) now has complete freedom to match the observed values. This means the function will be minimised when $g(x_{i}) = y_{i}$

```python
interpolate = sorted(list(zip(X, Y)), key=lambda x: x[0])
interpolate_x = list(map(lambda x: x[0], interpolate))
interpolate_y = list(map(lambda x: x[1], interpolate))
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.set_title('Generated data, true generating function g(X), and g hat')
ax.scatter(X, Y)
ax.plot(x_r, g_x)
ax.plot(interpolate_x, interpolate_y)
```
